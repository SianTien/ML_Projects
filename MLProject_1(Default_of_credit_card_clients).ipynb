multicolinearity *check*

```python
# for checking the multicolinearity we have to check the vif value
```

```python
import statsmodels.formula.api as sm
def cal_vif(input_data):
    x_vars = input_data
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]]
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.ols(formula="y~x", data=x_vars).fit().rsquared
        vif=round(1/(1-rsq),2)
        print (xvar_names[i], " VIF = " , vif)
```

vif for multicolinearity

```python
import statsmodels.formula.api as sm
#only x_vars should be given as input
def vif_func(input_data):
  x_vars=input_data
  feature_names=x_vars.columns
  for i in range(0,len(feature_names)):
    y=x_vars[feature_names[i]]
    x=x_vars[feature_names.drop(feature_names[i])]
    rsquare=sm.ols("y~x", data=x_vars).fit().rsquared
    vif=round(1/(1-rsquare),2)
    print("Feature Name ",feature_names[i], "VIF value is ", vif)
```

Oversampling and undersampling


```python
final_data=X_train_final
final_data["y"]=y_train
```

```python
count_of_good=len(final_data[final_data['y']==0])
count_of_bads=len(final_data[final_data['y']==1])
```

```python
count_of_bads
```

```python
count_of_good
```

```python
# for over and undersmapling we see that the count of bads are less so we twice it and take double of this and goods are more so we take 0.7 or 70% OF it
final_data_bads=final_data[final_data["y"]==1]
over_sample=final_data_bads.sample(count_of_bads*2, replace=True)
final_data_good=final_data[final_data["y"]==0]
under_sample=final_data_good.sample(int(count_of_good*0.7))
```

```python
# now conncating both over and under smaple
balanced_sample=pd.concat([over_sample,under_sample], axis=0)
balanced_sample.shape

balanced_sample["y"].value_counts()

X_balanced=balanced_sample.drop(["y"], axis=1)
y_balanced=balanced_sample["y"]
y_balanced.value_counts()
```

```python
# now we create the model on balanced data which we have created above
```
